```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime, time

import modutils

#data_dir = '../DataSets/Churn/'
src_dir = '../DataSets/kkbox/churn/raw/'
dst_dir = '../DataSets/kkbox/churn/proc/'


src_template = 'user_logs/uldt_{0}.csv'
proc_uids = 'uid_msno.csv'
dst_template = 'logs/uldtp_{0}.csv'

#t_years = ['2015', '2016']
t_years = ['2016']
t_months = ['01','02','03','04','05','06','07','08','09','10','11','12']
t_2017 = ['201701','201702','201703']
t_files = t_2017#[x+y for x in t_years for y in t_months]
cols = ['uid','date','num_25','num_50','num_75','num_985','num_100','num_unq','total_secs']
```

```python
uids_dict = pd.read_csv(dst_dir + proc_uids)
```

```python
t_files = [x for x in t_files if x > '201606']
```

```python
t_files
```

```python
print_format = 'Processed file ({0}) in {1:.2f} seconds (I: {2:.2f}, S: {3:.2f}, M:{4:.2f}, O: {5:.2f}), diff={6}'
for t_file in t_files:
    t0 = time.perf_counter()
    src = pd.read_csv(src_dir + src_template.format(t_file))
    src_len0 = len(src)
    t1 = time.perf_counter()
    src = src.sort_values(by=['msno','date'])
    t2 = time.perf_counter()
    src = pd.merge(left=src, right=uids_dict, how='inner', on='msno')
    t3 = time.perf_counter()
    src_len1 = len(src)
    src.to_csv(dst_dir + dst_template.format(t_file), index=False, columns=cols, float_format='%g')
    t4 = time.perf_counter()
    print(print_format.format(t_file, t4-t0,t1-t0,t2-t1,t3-t2,t4-t3,src_len1-src_len0))
```
