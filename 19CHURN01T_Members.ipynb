```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime, time

import modutils

#data_dir = '../DataSets/Churn/'
src_dir = '../DataSets/kkbox/churn/raw/'
dst_dir = '../DataSets/kkbox/churn/proc/'
data_fn = 'members_v3.csv'

dst_members = 'members.csv'
dst_uids = 'uids.csv'
```

```python
src = pd.read_csv(src_dir+data_fn)
```

```python
src = src.sort_values('msno')
```

```python
src[:10]
```

```python
src = src.reset_index(drop=True)
src['uid'] = src.index
```

```python
src[:10]
```

```python
t0 = time.perf_counter()
src.to_csv(dst_dir + dst_members, index=False)
t1 = time.perf_counter()
print('File written in {0:.2f} sec'.format(t1-t0))
```

```python
t0 = time.perf_counter()
src.to_csv(dst_dir + dst_uids, index=False, columns=['msno', 'uid'])
t1 = time.perf_counter()
print('File written in {0:.2f} sec'.format(t1-t0))
```
