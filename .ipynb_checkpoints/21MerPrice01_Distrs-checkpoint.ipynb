```python
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
import datetime, time

import modutils

data_dir = '../DataSets/MercariPrice/'
src_file = data_dir + 'train.csv' 
```

```python
src = pd.read_csv(src_file, '\t')
```

```python
src['cat1'] = src.category_name.map(lambda x: x.split('/')[0] if type(x) is str and len(x.split('/')) > 0 else 'NONE')
src['cat2'] = src.category_name.map(lambda x: x.split('/')[1] if type(x) is str and len(x.split('/')) > 1 else 'NONE')
src['cat3'] = src.category_name.map(lambda x: x.split('/')[2] if type(x) is str and len(x.split('/')) > 2 else 'NONE')
src['cat4'] = src.category_name.map(lambda x: x.split('/')[3] if type(x) is str and len(x.split('/')) > 3 else 'NONE')
src['cat5'] = src.category_name.map(lambda x: x.split('/')[4] if type(x) is str and len(x.split('/')) > 4 else 'NONE')
src['cat1g'] = src.cat1
src['cat2g'] = src.cat1g + '/' + src.cat2
src['cat3g'] = src.cat2g + '/' + src.cat3
```
