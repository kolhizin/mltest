```python
def get_transition_stats(df, var):
    df['nxt_'+var] = df[var].shift(-1).fillna(-1)
    df['nxt_'+var].view()[df.is_last==1] = -1
    df['nxt_'+var] = df['nxt_'+var].astype(type(df[var][0]))
    return df.groupby([var, 'nxt_'+var]).agg({'msno':'count', 'is_churn':'mean'}).reset_index().sort_values('msno', ascending=False)
```

```python
g = get_transition_stats(sr, 'is_auto_renew')
g
```

```python
var = 'is_cancel'
g = get_transition_stats(sr, var)
g['is_equal'] = 1*(g[var]==g['nxt_'+var])
g['is_last'] = 1*(g['nxt_'+var]==-1)
g['num_churn'] = g.is_churn * g.msno
g2 = g.groupby(['is_last','is_equal']).agg({'msno':'sum','num_churn':'sum'}).reset_index().sort_values('msno', ascending=False)
g2['churn'] = g2.num_churn / g2.msno
g2['share'] = g2.msno / g2.msno.sum()
g2
```
